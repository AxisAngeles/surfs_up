```python
# Dependencies
import numpy as np
import pandas as pd

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()

# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
inspector = inspect(engine)
columns = inspector.get_columns('measurement')
[print(column) for column in columns]
```

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
results = session.query(Measurement.date, Measurement.tobs).\
filter(func.extract("month",Measurement.date)==6).all()
```

```python
#  3. Convert the June temperatures to a list.
june_tobs = [row[1] for row in results]
type(june_tobs)
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
june_tobs_df = pd.DataFrame(june_tobs)
june_tobs_df.columns = ["June Temps"]
june_tobs_df.head()
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
june_tobs_df.describe()
```

## D2: Determine the Summary Statistics for December

```python
# 6. Write a query that filters the Measurement table to retrieve the temperatures for the month of December.
results = session.query(Measurement.date, Measurement.tobs).\
filter(func.extract("month",Measurement.date)==12).all()
```

```python
# 7. Convert the December temperatures to a list.
dec_tobs = [row[1] for row in results]
type(dec_tobs)
```

```python
# 8. Create a DataFrame from the list of temperatures for the month of December. 
dec_tobs_df = pd.DataFrame(dec_tobs)
dec_tobs_df.columns = ["December Temps"]
dec_tobs_df.head()
```

```python
# 9. Calculate and print out the summary statistics for the Decemeber temperature DataFrame.
dec_tobs_df.describe()
```
